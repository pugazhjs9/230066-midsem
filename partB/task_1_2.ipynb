# Task 1.2 — Key Assumptions

**Paper**:
 Efficient Variable Selection in Support Vector Machines via the Alternating Direction Method of Multipliers
(ICML 2011)

---

# Assumption 1

Assumption:

The method assumes that a sparse representation of the classifier weights exists, meaning that only a subset of the features are truly relevant for predicting the labels.

Why the method needs it:

The algorithm uses an $L_1$ regularization term $ \lambda \|w\|_1 $ in the optimization objective to enforce sparsity. This assumption is important because the effectiveness of $L_1$ regularization depends on the presence of irrelevant or weak features that can be driven to zero.

Violation scenario:

If most features in the dataset are genuinely important for prediction, forcing sparsity through $L_1$ regularization may remove useful information and degrade classification performance.

Paper reference:

The sparsity assumption appears in the formulation of the $L_1$-regularized SVM objective and the discussion of feature selection in the method section.

---

# Assumption 2

Assumption:

The optimization problem can be effectively decomposed into smaller subproblems that can be solved using the Alternating Direction Method of Multipliers (ADMM).

Why the method needs it:

The proposed algorithm relies on ADMM to split the original optimization problem into simpler updates for the variables $w$, $z$, and $u$. This decomposition is necessary for making the optimization scalable and computationally efficient.

Violation scenario:

If the problem structure does not allow effective decomposition or if the subproblems become computationally expensive, the benefits of ADMM may disappear and the optimization could become slower than traditional methods.

Paper reference:

The ADMM formulation and augmented Lagrangian discussion in the algorithm section of the paper.

---

# Assumption 3

Assumption:

The classification boundary can be adequately modeled using a linear SVM.

Why the method needs it:

The proposed method focuses on solving a linear SVM with feature selection. If the underlying relationship between features and labels is approximately linear, the classifier can learn an effective separating hyperplane.

Violation scenario:

If the dataset contains strongly nonlinear decision boundaries that cannot be captured by a linear model, the classifier may perform poorly even if the optimization algorithm converges successfully.

Paper reference:

The formulation of the linear SVM objective and the algorithm description in the paper.
